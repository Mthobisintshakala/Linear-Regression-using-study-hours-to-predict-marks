# Building a Linear regression that will predict Student Marks Base on their Study Hours

```python
import pandas as pd
study_marks = pd.read_csv(r"Downloads\\Study_hours_vs_Score_data.csv")
study_marks.head(2)
```

```python
study_marks.isnull().sum()
```

```python
study_marks.shape
```

```python
## Selecting dependent and independent variebles
x = study_marks[['Study_hours']]
y = study_marks[['Final_Marks']]
```

```python
# spliting the dataset into Training and testing data in ratio of 75:25
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, train_size = 0.75)
```

```python
#Creating a study_marks_model using the LinearRegression
from sklearn.linear_model import LinearRegression
study_marks_model = LinearRegression()
study_marks_model
```

```python
#Fitting the study_markd_model using the training data
study_marks_model.fit(x_train, y_train)
```

```python
# Makes Predictions using the study_marks_model to predict the marks of the student using th testing data
marks_predicted = study_marks_model.predict(x_test)
marks_predicted
```

```python
# Evaluating the study_marks_model using r2_score using the testing data
from sklearn.metrics import r2_score
study_marks_score = r2_score(marks_predicted, y_test)
study_marks_score
```

```python
# ploting the praph to visualise the study_marks_model Performance
import matplotlib.pyplot as plt
import seaborn as sb
plt.scatter( x_test, y_test, label='Original Testing Values', color='c', marker=5)
plt.plot(x_test,  marks_predicted, label='Predicted values', color='red')
plt.xlabel('Hours Spend by Student To Study')
plt.ylabel('Marks Obtained by Student')
plt.legend()
plt.show()
```
